## Nifty-50 Stock Price Prediction Tool

### Disclaimer: This program is just a learning tool to understand deep learning using LSTM layers. Do not take the predictions from this model as accurate financial advice. Please use your discretion and invest mindfully.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import tensorflow as tf
```

Collecting data from the csv files in the data folder

```python
import os, glob
filenames = []
for file in glob.glob("data/*.csv"):
    filenames.append(file)

filenames.sort()
filenames = filenames[:-1]
# filenames
```

```python
num_stocks = len(filenames)
stocks = [filenames[i][5:-4] for i in range(num_stocks)]
stocks
```

Getting just the stock prices at "Close" of day and storing them into a dataframe. For our prediction purpose, it is enough to choose the stock prices in the past 1000 days (in the interest of time and accuracy)

```python
data_dict = {}
num_days = 1000
for i in range(num_stocks):
  stock_dat = pd.read_csv(filenames[i])
  data_dict[stocks[i]] = stock_dat.tail(num_days).reset_index()['Close']
```

```python
data = pd.DataFrame.from_dict(data_dict)
data.head()
```

```python
plt.plot(data[stocks[0]])
plt.title(stocks[0])
plt.xlabel('Time')
plt.ylabel('Stock Price')
```

Scaling stock price data individually 

```python
#First time run
from sklearn.preprocessing import MinMaxScaler
scalers = {stocks[i]: MinMaxScaler() for i in range(num_stocks)}
for i in range(num_stocks):
  data[stocks[i]] = scalers[stocks[i]].fit_transform(np.array(data[stocks[i]]).reshape(num_days,1))
```

```python
data.head()
```

Train-test split the time series stock price data

```python
train_frac = 0.7

train_size = int(train_frac*num_days)
test_size = num_days - train_size

data_train,data_test = data.iloc[:train_size],data.iloc[train_size:]
```

```python
data_train.shape,data_test.shape
```

### Model explanation

To predict the stock price on the 101th day, we input the stock prices of the previous 100 days (day 1 to day 100) into the model to get the model predicted output for the 101st day. 

Similarly for the 102th day prediction, we input the previous 100 days (day 2 to day 101) stock prices into the model to get the output for 102nd day... and so on.

Hence the train data and test data are designed as stock price groups of 100:<br />
<p>data 1 - X: day1 to day100 stock price; y: day101 stock price</p>
<p>data 2 - X: day2 to day101 stock price; y: day102 stock price</p>
.<br />
.<br />
.<br />

which will go into the model as input

`time_step` value can be modified to change the group size

```python
def create_X_and_y(data,time_step=1):
  dataX, datay = {stocks[i]: [] for i in range(num_stocks)}, {stocks[i]: [] for i in range(num_stocks)}
  for i in range(len(data)-time_step-1):
    for j in range(num_stocks):
      dataX[stocks[j]].append(data[stocks[j]].iloc[i:(i+time_step)])
      datay[stocks[j]].append(data[stocks[j]].iloc[(i+time_step)])
  
  return dataX, datay
```

```python
time_step = 100
X_train, y_train = create_X_and_y(data_train,time_step=time_step)
X_test, y_test = create_X_and_y(data_test,time_step=time_step)
```

```python
x_train_shape, y_train_shape = np.array(X_train[stocks[0]]).shape, np.array(y_train[stocks[0]]).shape
```

```python
x_test_shape, y_test_shape = np.array(X_test[stocks[0]]).shape, np.array(y_test[stocks[0]]).shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
```

Deep Learning model is invidiually created for each stock 

```python
#Load models from files
models = {}
for i in range(num_stocks):
    models[stocks[i]] = tf.keras.models.load_model('models/'+stocks[i]+'_model.h5')
```

```python
def create_model():  
  model = Sequential()
  model.add(LSTM(100,return_sequences=True,input_shape=(time_step,1)))
  model.add(LSTM(50,return_sequences=True))
  model.add(LSTM(25))
  model.add(Dense(1))
  model.compile(loss='mean_squared_error',optimizer='adam')
  return model

models = {stocks[i]: create_model() for i in range(num_stocks)}
```

```python
models[stocks[0]].summary()
```

Model is fitted according to the input values. Run time ~ 1 hour (Core i5, 8GB RAM, no dedicated graphics card)

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)
history = {}
for i in range(num_stocks):
  xtrain = np.array(X_train[stocks[i]]).reshape(x_train_shape[0],x_train_shape[1],1)
  ytrain = np.array(y_train[stocks[i]]).reshape(y_train_shape[0],1)
  xtest = np.array(X_test[stocks[i]]).reshape(x_test_shape[0],x_test_shape[1],1)
  ytest = np.array(y_test[stocks[i]]).reshape(y_test_shape[0],1)
  history[stocks[i]] = models[stocks[i]].fit(xtrain,ytrain,
                                             validation_data=(xtest,ytest),
                                             epochs=100,
                                             batch_size=64,
                                             verbose=1,
                                             callbacks=[callback])
```

Using the model, we can predict the train and test days' stock price as well and storing the plots

```python
train_predict = {stocks[i]: [] for i in range(num_stocks)}
test_predict = {stocks[i]: [] for i in range(num_stocks)}

for i in range(num_stocks):
  xtrain = np.array(X_train[stocks[i]]).reshape(x_train_shape[0],x_train_shape[1],1)
  xtest = np.array(X_test[stocks[i]]).reshape(x_test_shape[0],x_test_shape[1],1)

  train_predict[stocks[i]] = models[stocks[i]].predict(xtrain)
  test_predict[stocks[i]] = models[stocks[i]].predict(xtest)

  train_predict[stocks[i]] = np.array(scalers[stocks[i]].inverse_transform(train_predict[stocks[i]])).reshape(-1)
  test_predict[stocks[i]] = np.array(scalers[stocks[i]].inverse_transform(test_predict[stocks[i]])).reshape(-1)
```

```python
def train_test_pred(stock_key):
    curr_stock = stocks[stock_key]

    fig,[ax1,ax2] = plt.subplots(ncols=2,nrows=1,figsize=(14,5))
    ax1.plot(train_predict[curr_stock])

    temp_train = np.array(data[curr_stock].iloc[time_step:len(train_predict[curr_stock])+time_step]).reshape(len(train_predict[stocks[0]]),1)
    temp_train = np.array(scalers[curr_stock].inverse_transform(temp_train)).reshape(-1)
    ax1.plot(temp_train)

    ax1.legend(['Predicted','Actual'])
    ax1.set_title('Train')
    ax1.set_xlabel('Days')
    ax1.set_ylabel('Stock Price')
    ax1.grid(True)

    ax2.plot(test_predict[curr_stock])

    temp_test = np.array(data[curr_stock].iloc[len(temp_train)+2*time_step+1:num_days-1]).reshape(len(test_predict[stocks[0]]),1)
    temp_test = np.array(scalers[curr_stock].inverse_transform(temp_test)).reshape(-1)
    ax2.plot(temp_test)

    ax2.legend(['Predicted','Actual'])
    ax2.set_title('Test')
    ax2.set_xlabel('Days')
    ax2.set_ylabel('Stock Price')
    ax2.grid(True)
    
    fig.suptitle(curr_stock)
    
    fname = 'static/images/'+curr_stock+'_train_test.png'
    fig.savefig(fname)
    plt.close()

for i in range(num_stocks):
    train_test_pred(i)
```

```python
def together_pred(stock_key):
    curr_stock = stocks[stock_key]
    fig,ax = plt.subplots(figsize=(7,5))
    
    temp_train = np.array(data[curr_stock].iloc[time_step:len(train_predict[curr_stock])+time_step]).reshape(len(train_predict[stocks[0]]),1)
    temp_test = np.array(data[curr_stock].iloc[len(temp_train)+2*time_step+1:num_days-1]).reshape(len(test_predict[stocks[0]]),1)

    train_pred_plot = np.empty_like(data[curr_stock])
    train_pred_plot[:] = np.nan
    train_pred_plot[time_step:len(train_predict[stocks[0]])+time_step] = train_predict[curr_stock]


    test_pred_plot = np.empty_like(data[curr_stock])
    test_pred_plot[:] = np.nan
    test_pred_plot[len(temp_train)+2*time_step+1:num_days-1] = test_predict[curr_stock]

    ax.plot(scalers[curr_stock].inverse_transform(np.array(data[curr_stock]).reshape(num_days,1)))
    ax.plot(train_pred_plot)
    ax.plot(test_pred_plot)
    ax.legend(['Actual','Predicted (train)','Predicted (test)'])
    ax.set_title(curr_stock)
    ax.set_xlabel('Days')
    ax.set_ylabel('Stock Price')
    ax.grid(True)
    
    fname = 'static/images/'+curr_stock+'_together.png'
    fig.savefig(fname)
    plt.close()

for i in range(num_stocks):
    together_pred(i)
```
